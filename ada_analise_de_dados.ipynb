# Aula 04
Outlier

```python
houses = pd.read_csv('/content/houses_to_rent.csv')
```

```python
houses.head()
```

```python
houses.describe()
```

```python
houses['bathroom'].value_counts()
```

```python
# outlier
# critetrio IQR > DISTANCIA INTER QUARTIL

q1 = houses['bathroom'].quantile(0.25) ## primerio qurtil
q3 = houses['bathroom'].quantile(0.75)

iqr = q3 - q1

print(f"IQR: {iqr}")
```

```python
houses_outlier = houses[(houses['bathroom'] < (q1 - 1.5 * iqr)) | (houses['bathroom'] > (q3 + 1.5 * iqr))]
```

```python
houses_outlier
```

```python
# tratar Outierl
# remove

houses_inliers = houses[(houses['bathroom'] >= q1 - (1.5 * iqr)) & (houses['bathroom'] <= q3 + (1.5 * iqr))]
```

```python
houses_inliers.head()
```

```python
houses_inliers.describe()
```

```python
ax = candy_filtred.plot(x='observation_date', y='industrial_production', figsize=(12, 6))
coords = ['2011-01-01', '2012-01-01', '2013-01-01', '2014-01-01', '2015-01-01', '2016-01-01', '2017-01-01']

for xc in coords:
  plt.axline(x=xc, color='black', lineStyles='--')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

candy_filtred.set_index('observation_date', inplace=True)
analysis = candy_filtred[['industrial_production']].copy()

decompose_result = seasonal_decompose(analysis, model='multiplicative')

trend = decompose_result.trend
seasonal = decompose_result.seasonal
residual = decompose_result.resid

decompose_result.plot()
```
